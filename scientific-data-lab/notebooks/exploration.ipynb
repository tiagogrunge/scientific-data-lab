# Scientific Data Lab - Exploration Notebook

This notebook demonstrates the complete pipeline of the scientific-data-lab project:
1. Load experimental data from CSV
2. Apply signal processing (normalization and moving average)
3. Calculate statistical measures
4. Visualize the results

## Step 1: Import Libraries

Import all necessary modules from our project and external libraries.

```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Add src to path to import our modules
sys.path.insert(0, '../')

from src.data_loader import load_csv
from src.signal_processing import moving_average, normalize
from src.analysis import calculate_statistics
from src.visualization import plot_signal

print("All libraries imported successfully!")
```

## Step 2: Load Experimental Data

Load the sample CSV file containing time and signal columns.

```python
# Load the sample data
df = load_csv('../sample_data/experiment_001.csv')

print(f"Data loaded successfully!")
print(f"Shape: {df.shape}")
print(f"\nFirst few rows:")
print(df.head())
print(f"\nData info:")
print(df.info())
```

## Step 3: Extract and Plot Raw Signal

Extract the signal column and visualize the raw experimental data.

```python
# Extract the signal data
raw_signal = df['signal'].values

print(f"Raw signal shape: {raw_signal.shape}")
print(f"Raw signal range: [{raw_signal.min():.2f}, {raw_signal.max():.2f}]")

# Plot raw signal
plot_signal(raw_signal, title="Raw Experimental Signal")
```

## Step 4: Apply Signal Processing

Apply normalization and moving average filtering to smooth the signal.

```python
# Normalize the signal
normalized_signal = normalize(raw_signal)

print("Normalized signal statistics:")
print(f"  Mean: {normalized_signal.mean():.6f}")
print(f"  Std: {normalized_signal.std():.6f}")
print(f"  Min: {normalized_signal.min():.4f}")
print(f"  Max: {normalized_signal.max():.4f}")

# Apply moving average filter
window_size = 3
smoothed_signal = moving_average(normalized_signal, window_size=window_size)

print(f"\nApplied moving average with window size: {window_size}")
```

## Step 5: Visualize Processing Results

Compare raw, normalized, and smoothed signals.

```python
# Create a comparison plot
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

# Plot raw signal
axes[0].plot(raw_signal, 'b-', linewidth=2)
axes[0].set_title('Raw Signal', fontsize=12, fontweight='bold')
axes[0].set_ylabel('Signal Value')
axes[0].grid(True, alpha=0.3)

# Plot normalized signal
axes[1].plot(normalized_signal, 'g-', linewidth=2)
axes[1].set_title('Normalized Signal (zero mean, unit variance)', fontsize=12, fontweight='bold')
axes[1].set_ylabel('Signal Value')
axes[1].grid(True, alpha=0.3)

# Plot smoothed signal
axes[2].plot(raw_signal, 'b--', alpha=0.5, linewidth=1.5, label='Raw')
axes[2].plot(smoothed_signal, 'r-', linewidth=2, label='Smoothed')
axes[2].set_title(f'Moving Average Filter (window={window_size})', fontsize=12, fontweight='bold')
axes[2].set_xlabel('Sample Index')
axes[2].set_ylabel('Signal Value')
axes[2].legend()
axes[2].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## Step 6: Calculate Statistics

Compute descriptive statistics for the processed signal.

```python
# Calculate statistics for each signal
raw_stats = calculate_statistics(raw_signal)
norm_stats = calculate_statistics(normalized_signal)
smooth_stats = calculate_statistics(smoothed_signal)

# Display results
print("=" * 60)
print("STATISTICAL ANALYSIS")
print("=" * 60)

print("\nRaw Signal Statistics:")
for key, value in raw_stats.items():
    print(f"  {key:8s}: {value:10.4f}")

print("\nNormalized Signal Statistics:")
for key, value in norm_stats.items():
    print(f"  {key:8s}: {value:10.4f}")

print("\nSmoothed Signal Statistics:")
for key, value in smooth_stats.items():
    print(f"  {key:8s}: {value:10.4f}")

print("\n" + "=" * 60)
```

## Step 7: Summary and Insights

Extract key insights from the analysis.

```python
print("ANALYSIS SUMMARY")
print("="*60)
print(f"\n✓ Data loaded: {len(df)} samples")
print(f"✓ Signal range (raw): {raw_stats['min']:.2f} to {raw_stats['max']:.2f}")
print(f"✓ Signal mean (raw): {raw_stats['mean']:.4f}")
print(f"✓ Signal std (raw): {raw_stats['std']:.4f}")
print(f"\n✓ After normalization:")
print(f"    Mean: {norm_stats['mean']:.6f} (should be ~0)")
print(f"    Std: {norm_stats['std']:.6f} (should be ~1)")
print(f"\n✓ After smoothing:")
print(f"    Noise reduction: {(1 - smooth_stats['std']/raw_stats['std'])*100:.1f}% std reduction")
print(f"\nNext steps:")
print(f"  → Try different window sizes for moving average")
print(f"  → Implement additional filters (e.g., Butterworth)")
print(f"  → Analyze frequency content with FFT")
print(f"  → Compare with other smoothing algorithms")
print("\n" + "="*60)
```
